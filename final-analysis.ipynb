```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
```

```python
data = pd.read_csv("./utils/mitsubishi_imiev.csv", encoding='latin-1')
avg_speed = pd.read_csv('./utils/volkswagen_e_golf.csv')
avg_speed
data.shape
```

```python
data.drop('avg_speed(km/h)', axis=1, inplace=True)
a = avg_speed["avg_speed(km/h)"].loc[0:1285]
data = pd.concat([data, a], axis=1)
data
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.drop(['manufacturer', 'model', 'fuel_type', 'version', 'fuel_note', 'power(kW)'], axis=1, inplace=True)
```

```python
data.dtypes
```

```python
data.fuel_date = pd.to_datetime(data.fuel_date)
```

```python
data
```

```python
data.fuel_date = pd.to_numeric(pd.to_datetime(data.fuel_date))
```

```python
data = pd.get_dummies(data, drop_first=True)
```

```python
from sklearn.model_selection import train_test_split
X = data.drop('consumption(kWh/100km)', axis=1)
y = data["consumption(kWh/100km)"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) 
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test  = scaler.fit_transform(X_test)

```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(X_train, y_train)

predictions = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
print(r2_score(y_test, predictions))
print(mean_squared_error(y_test, predictions))
```

```python
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf')
regressor.fit(X_train, y_train)

svm_preds = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
print(r2_score(y_test, svm_preds))
print(mean_squared_error(y_test, svm_preds))
```

```python
from sklearn.linear_model import SGDRegressor
reg = SGDRegressor()
reg.fit(X_train, y_train)
sgd_preds = reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
print(r2_score(y_test, sgd_preds))
print(mean_squared_error(y_test, sgd_preds))
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor()
knn.fit(X_train, y_train)
knn_preds = knn.predict(X_test)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
print(r2_score(y_test, knn_preds))
print(mean_squared_error(y_test, knn_preds))
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import DotProduct, WhiteKernel
kernel = DotProduct() + WhiteKernel()
gpr = GaussianProcessRegressor(kernel=kernel)
gpr.fit(X_train, y_train)
gpr_preds = reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
print(r2_score(y_test, gpr_preds))
print(mean_squared_error(y_test, gpr_preds))
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
dtr.fit(X_train, y_train)
dtr_preds = dtr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
print(r2_score(y_test, dtr_preds))
print(mean_squared_error(y_test, dtr_preds))
```
